```python
fd = open("products.json",'r')
r = fd.read()
fd.close()

products = json.loads(r)
```

```python
prod_id = input("Enter the product Id: ")
quan_no = int(input("Enter the quantity of the Product: "))

for i in products.keys():
    while prod_id == i:
        if quan_no <= products[i]['quan']:
            print("_________________________________________________")
            print("product:",products[prod_id]['name'])
            print("category:",products[prod_id]['category'])
            print("price:",products[prod_id]['price'])
            print("Expiry date:",products[prod_id]['expiry'])
            print("Brand:",products[prod_id]['brand'])
            print("_________________________________________________ ")
            print("----------BILLING DETAILS:---------- ")
            print("Total Amount: ",products[prod_id]['price']*quan_no) 
            print("Enjoy Your Day")
            print("Thank You. Visit Again")
            print("__________________________________________________")
            products[prod_id]["quan"] = products[prod_id]["quan"] - quan_no
        else:
            print("Sorry. Stock of product you enterd is: ",products[prod_id]['quan'],"\nSorry for the inconvenience")
        break
if prod_id not in products.keys():
   print("Sorry. No product found : ",prod_id,"\Please enter a valid product ID")

       
js = json.dumps(products)

fd = open("products.json",'w')
fd.write(js)
fd.close()



fd = open("sales.json",'r')
fd.read()
fd.close()

sales = json.loads(r)
num = len(sales)+1

sales[num] = {'prod':prod_id, 'quan' : quan_no, 'amount': products[prod_id]['price'] * quan_no}
sale = json.dumps(sales)

fd = open("sales.json",'w')
fd.write(sale)
fd.close()


fd = open("sales.json",'w')
fd.write(sale)
fd.close()
```
